Bagging method is assembling multiple algorithms and see which give sthe Best result.
Random Forest is one of the Example of Bagging Method,where multiple Decision Trees are combined.estimators-> this parameter specifies the number of decision Trees

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
X = dataset.data
y = dataset.target
```

```python
#Split the data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y)
```

```python
# 3️⃣ Feature scaling (VERY IMPORTANT for KNN)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#create a Model of KNeighbors Classifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
y_pred=knn.predict(X_test)
```

```python
y_pred
```

```python
#create a Model of Bagging Classifier
bag_knn=BaggingClassifier(knn,n_estimators=10)
bag_knn.fit(X_train,y_train)
bag_knn.score(X_test,y_test)
```
