Import the required Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Create a dummy data to understand Data Imputation flow


```python
data={'age':[25,np.nan,30,np.nan,35],
      'salary':[50000,60000,np.nan,90000,np.nan]

      }
```

```python
dataframe=pd.DataFrame(data)
```

```python
dataframe
```

Data Imputation done via Mean,Meadian,Mode

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')#for Outliers strategy='median',for Categorical data strategy='mode'(hihest occurance value)
imputed_data=imputer.fit_transform(dataframe)
imputed_df= pd.DataFrame(imputed_data,columns=dataframe.columns)
```

```python
print(imputed_df)
```

data Imputation Via KNN Imputer

```python
from sklearn.impute import KNNImputer
knn_imputer=KNNImputer(n_neighbors=2)
knn_imputed_data=knn_imputer.fit_transform(dataframe)
knn_imputed_df= pd.DataFrame(knn_imputed_data,columns=dataframe.columns)
```

```python
print(knn_imputed_df)
```
