Import required libraries


```python
import numpy as np
import pandas as pd
```

```python
data=np.array([[3,7],
               [-4,-6],
               [1,-1],
               [7,8],
               [-4,-1],
               [-3,-7]])
```

```python
dataframe =pd.DataFrame(data,columns=['x1','x2'])
```

```python
dataframe
```

Next steps:

1.To check whether Data is in Normal Standard form

Standard Normal Form : A data is said to be in Normal Standard for if Mean =0 and Standard Deviation =1


```python
#to check if the data is in the Normal Standard form
dataframe.describe()
```

```python
#Since we see that mean is not Eqaul to 1 ,we do the preproceesing to get the Normal Data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
datascaled=scaler.fit_transform(dataframe)
dataframe =pd.DataFrame(datascaled,columns=['x1','x2'])
```

```python
dataframe.describe()
```

2.Correlation between the Matrix that we have

Difference between Covariance matrix Vs Correlation matrix
--+ve Covariance -> if F1 increases  then F2 also increases
---ve Covariance -> if F1 decreases then F2 increses or Vice versa
---0  F1 and F2 are not in relationship with each other
===Covraince does not have any specific Range====

------------Correlation matrix-----------------
--- -ve Correlation -> indicates the value is close to -1
--- +ve Correlation -> indicates the value is close to +1

Covariance Matrix  gives the accurate Value(it is scaled value)
Correlation Matrix gives the Strength of the Relation ship


**Creating the Coraviance matrix**

```python
#Approach #1
c1=dataframe.x1
c2=dataframe.x2
np.cov(c1,c2) #Evaluate the Covariance matrix
```

```python
#approach #2
covarianceMatrix=dataframe.T @ dataframe / 5 #(here 5= #of datapoints - 1 )
```

```python
covarianceMatrix
```

```python
#Approach #3
np.sum(c1 * c2)/5
```

**Step 3.Evaluate the Eigen Values and Eigen Vectors of the above Correlation Matrix

Using eigenValues and eigenVectors Principal Components are created
1.Higest eigenValue indicates the highest Information

Example :- PC1 will be with the highest eigenValue**

```python
eigenValues,eigenVectors=np.linalg.eig(covarianceMatrix)
```

```python
eigenValues
```

```python
eigenVectors
```

```python
eigenVectors[:,0] #0th column
```

```python
eigenVectors[:,1]#1st Column
```

```python
#PC1 contains the mximum information of the original two features that you have
PC1 =dataframe @ eigenVectors[:,0]
PC1
```

```python
PC2 =dataframe @ eigenVectors[:,1]
PC2
```
