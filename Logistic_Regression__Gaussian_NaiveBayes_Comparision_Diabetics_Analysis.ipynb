```python
##Outliers Detection
##Split the Data into Input Features and Output features

X=dataframe.drop(columns='Outcome',axis=1)
y=dataframe['Outcome']
```

```python
X.shape
```

```python
y.shape
```

```python
##boxplot method is used to detect the Outliers
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=X,ax=ax)
plt.savefig('boxplot.jpg')
```

```python
#Inter Quantile Range method --- Approach1 to remove Outliers
```

```python
##Identifying Outliers for multiple Columns
cols=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
for col in cols:
  Q1=X[col].quantile(0.25)
  Q3=X[col].quantile(0.75)
  IQR= Q3 - Q1
  lower_bound= Q1 - 1.5 * IQR
  upper_bound= Q3 + 1.5 * IQR
  mask=(X[col]>=lower_bound) & (X[col]<=upper_bound)
```

```python
X_outlier_detection=X[mask]
y_outlier_detection=y[mask]
```

```python
X_outlier_detection.shape
```

```python
y_outlier_detection.shape
```

```python
##Standardization
#Standard Normal Form ---->Mean =0 and Standard Deviation =1
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_outlier_detection)
```

```python
##boxplot method is used to detect the Outliers after Standardization
fig,ax=plt.subplots(figsize=(15,15))
sns.boxplot(data=X_scaled,ax=ax)
plt.savefig('boxplot-std.jpg')
```

```python
##Remove outliers using Quantiles
q=X_scaled['Insulin'].quantile(.95) #try for other Quantile also
mask=X_scaled['Insulin']<q
datanew=X_scaled[mask]
y_outlier_detection=y_outlier_detection[mask]
```

```python
##Splitting the Data into Training an Testing
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(datanew,y_outlier_detection,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

**This gives number of Class Counts.Since there is lot of differenc e,this is an imbalanced data**

```python
y_train.value_counts()
```

**Imbalance Data can be handled by
--Over-Sampling
--Undersampling  
--SMOTE
OverSampling - Take the Minority Class and Increase that Number to the Mjority class

UnderSampling - Take the Majority Class and Decrease that Number to the Minority class

SMOTE - Take Synthetic Class and increase number of samples to the Majority class
**

```python
#SMOTE Technique
from imblearn.over_sampling import SMOTE
smote =SMOTE(random_state=42)
X_train_resampled,y_train_resampled=smote.fit_resample(X_train,y_train)

#Check the resampled class distribution
print("\nResampledClass Distribution")
print(pd.Series(y_train_resampled).value_counts())

```

```python
from sklearn.linear_model import LogisticRegression
classification=LogisticRegression()
classification.fit(X_train_resampled,y_train_resampled)
```

```python
y_predictions=classification.predict(X_test)
print(y_predictions)
```

```python
#Model Evaluation
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy=accuracy_score(y_test,y_predictions)
```

```python
print("Accuracy of the Model:",accuracy)
```

```python
target_names=['Non-Diabetic','Diabetic']
```

```python
print(classification_report(y_test,y_predictions,target_names=target_names))
```

**Gaussian Naive Bayes Algorithm Implementation**

```python
from sklearn.naive_bayes import GaussianNB
model_gaussian_naive_bayes=GaussianNB()
model_gaussian_naive_bayes.fit(X_train_resampled,y_train_resampled)
```

```python
y_predict_gaussian_naive_bayes = model_gaussian_naive_bayes.predict(X_test)
print(y_predict_gaussian_naive_bayes)
```

```python
print("Confusion matrix")
print(confusion_matrix(y_test,y_predict_gaussian_naive_bayes))
```

```python
print("Classification Report")
print(classification_report(y_test,y_predict_gaussian_naive_bayes))
```

```python
print("Accuracy Report")
accuracy=accuracy_score(y_test,y_predict_gaussian_naive_bayes)
print(accuracy)
```
