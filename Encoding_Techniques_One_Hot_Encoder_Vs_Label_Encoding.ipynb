```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
#Load the Dataset
file_path = '/content/carprices.csv'
car_data = pd.read_csv(file_path)
```

```python
#Dispaly first few rows of the Dataset to understand the data
car_data.head()
```

```python
car_data.shape
```

```python
#Extract Car Models
car_models=car_data['Car Model']
car_models
```

```python
#Apply One Hot Encoding
one_hot_encoder=OneHotEncoder(sparse_output=False)
one_hot_encoded=one_hot_encoder.fit_transform(car_models.values.reshape(-1,1))

```

```python
one_hot_encoded
```

```python
#Convert One_hot_encoding into DataFrame
one_hot_encoded_df=pd.DataFrame(one_hot_encoded,columns=one_hot_encoder.get_feature_names_out(['Car Model']))
```

```python
#Combine One Hot Encoded Columns into the main Dataframe
car_data_one_hot_encoded=pd.concat([car_data,one_hot_encoded_df],axis=1)
```

```python
#Display the One Hot encoded Data
print("Car One hot encoded Data",car_data_one_hot_encoded)
```

```python
#Apply label Encoder of the Car data
label_encoder=LabelEncoder()
label_encoded=label_encoder.fit_transform(car_data['Car Model'])
```

```python
#Add the label Encoded Data to the Original Data
car_data_label_encoded=car_data.copy()
car_data['Car Model (Label Encoded)']=label_encoded
```

```python
#Display the Label Encoded Data
print("The Label Encoded Data",car_data_label_encoded.head(14))
```
